```
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

blood_sugar = ctrl.Antecedent(np.arange(50, 301, 1), 'blood_sugar')
urination = ctrl.Antecedent(np.arange(0, 11, 1), 'urination')
BMI = ctrl.Antecedent(np.arange(10, 41, 1), 'BMI')
fatigue = ctrl.Antecedent(np.arange(0, 11, 1), 'fatigue')

risk = ctrl.Consequent(np.arange(0, 101, 1), 'risk')

blood_sugar['low'] = fuzz.trimf(blood_sugar.universe, [50, 80, 120])
blood_sugar['normal'] = fuzz.trimf(blood_sugar.universe, [80, 120, 200])
blood_sugar['high'] = fuzz.trimf(blood_sugar.universe, [120, 200, 300])
blood_sugar['very_high'] = fuzz.trimf(blood_sugar.universe, [200, 300, 300])

urination['rare'] = fuzz.trimf(urination.universe, [0, 1, 3])
urination['sometimes'] = fuzz.trimf(urination.universe, [2, 5, 7])
urination['frequent'] = fuzz.trimf(urination.universe, [6, 9, 10])

BMI['underweight'] = fuzz.trimf(BMI.universe, [10, 15, 18.5])
BMI['normal'] = fuzz.trimf(BMI.universe, [18.5, 22, 25])
BMI['overweight'] = fuzz.trimf(BMI.universe, [25, 27, 30])
BMI['obese'] = fuzz.trimf(BMI.universe, [30, 35, 40])


fatigue['low'] = fuzz.trimf(fatigue.universe, [0, 1, 4])
fatigue['medium'] = fuzz.trimf(fatigue.universe, [3, 5, 7])
fatigue['high'] = fuzz.trimf(fatigue.universe, [6, 9, 10])


risk['low'] = fuzz.trimf(risk.universe, [0, 10, 30])
risk['medium'] = fuzz.trimf(risk.universe, [20, 50, 70])
risk['high'] = fuzz.trimf(risk.universe, [60, 80, 90])
risk['very_high'] = fuzz.trimf(risk.universe, [85, 100, 100])


rule1 = ctrl.Rule(blood_sugar['very_high'] & urination['frequent'], risk['very_high'])
rule2 = ctrl.Rule(blood_sugar['very_high'] & urination['sometimes'], risk['high'])
rule3 = ctrl.Rule(blood_sugar['high'] & urination['frequent'], risk['high'])
rule4 = ctrl.Rule(blood_sugar['normal'] & urination['rare'], risk['low'])
rule5 = ctrl.Rule(blood_sugar['very_high'] & BMI['obese'], risk['very_high'])
rule6 = ctrl.Rule(blood_sugar['very_high'] & BMI['overweight'], risk['high'])
rule7 = ctrl.Rule(blood_sugar['high'] & BMI['overweight'], risk['high'])
rule8 = ctrl.Rule(blood_sugar['high'] & BMI['normal'], risk['medium'])
rule9 = ctrl.Rule(blood_sugar['normal'] & BMI['normal'], risk['low'])
rule10 = ctrl.Rule(blood_sugar['very_high'] & fatigue['high'], risk['very_high'])
rule11 = ctrl.Rule(blood_sugar['high'] & fatigue['high'], risk['high'])
rule12 = ctrl.Rule(blood_sugar['high'] & fatigue['medium'], risk['medium'])
rule13 = ctrl.Rule(blood_sugar['normal'] & fatigue['low'], risk['low'])

risk_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13])
risk_simulation = ctrl.ControlSystemSimulation(risk_ctrl)


risk_simulation.input['blood_sugar'] = 250
risk_simulation.input['urination'] = 8
risk_simulation.input['BMI'] = 32
risk_simulation.input['fatigue'] = 9

risk_simulation.compute()
print(risk_simulation.output['risk'])

variables = [blood_sugar, urination, BMI, fatigue, risk]
for var in variables:
    for term in var.terms:
       plt.plot(var.universe, var[term].mf, label=term)
    plt.title(var.label)
    plt.legend()
    plt.show()
```

```
import numpy as np
import skfuzzy as fuzz
import skfuzzy.control as ctrl
import matplotlib.pyplot as plt

heart_rate = ctrl.Antecedent(np.arange(40, 161, 1), 'heart_rate')
blood_pressure = ctrl.Antecedent(np.arange(70, 181, 1), 'blood_pressure')
body_temp = ctrl.Antecedent(np.arange(34, 40, 0.1), 'body_temp')
oxygen_saturation = ctrl.Antecedent(np.arange(85, 101, 1), 'oxygen_saturation')
blood_sugar = ctrl.Antecedent(np.arange(50, 200, 1), 'blood_sugar')

drug_dose = ctrl.Consequent(np.arange(-1, 2, 1), 'drug_dose')  # -1: Reduce, 0: Maintain, 1: Increase
fluid_therapy = ctrl.Consequent(np.arange(-1, 2, 1), 'fluid_therapy')
oxygen_therapy = ctrl.Consequent(np.arange(-1, 2, 1), 'oxygen_therapy')

heart_rate['low'] = fuzz.trimf(heart_rate.universe, [40, 40, 60])
heart_rate['normal'] = fuzz.trimf(heart_rate.universe, [60, 80, 100])
heart_rate['high'] = fuzz.trimf(heart_rate.universe, [100, 130, 160])

blood_pressure['low'] = fuzz.trimf(blood_pressure.universe, [70, 70, 90])
blood_pressure['normal'] = fuzz.trimf(blood_pressure.universe, [90, 110, 130])
blood_pressure['high'] = fuzz.trimf(blood_pressure.universe, [130, 150, 180])

body_temp['low'] = fuzz.trimf(body_temp.universe, [34, 34, 36])
body_temp['normal'] = fuzz.trimf(body_temp.universe, [36, 36.5, 37.5])
body_temp['high'] = fuzz.trimf(body_temp.universe, [37.5, 38.5, 39.5])

oxygen_saturation['low'] = fuzz.trimf(oxygen_saturation.universe, [85, 85, 90])
oxygen_saturation['normal'] = fuzz.trimf(oxygen_saturation.universe, [90, 95, 100])

blood_sugar['low'] = fuzz.trimf(blood_sugar.universe, [50, 50, 70])
blood_sugar['normal'] = fuzz.trimf(blood_sugar.universe, [70, 100, 140])
blood_sugar['high'] = fuzz.trimf(blood_sugar.universe, [140, 160, 200])

drug_dose['reduce'] = fuzz.trimf(drug_dose.universe, [-1, -1, 0])
drug_dose['maintain'] = fuzz.trimf(drug_dose.universe, [-0.5, 0, 0.5])
drug_dose['increase'] = fuzz.trimf(drug_dose.universe, [0, 1, 1])

fluid_therapy['reduce'] = fuzz.trimf(fluid_therapy.universe, [-1, -1, 0])
fluid_therapy['maintain'] = fuzz.trimf(fluid_therapy.universe, [-0.5, 0, 0.5])
fluid_therapy['increase'] = fuzz.trimf(fluid_therapy.universe, [0, 1, 1])

oxygen_therapy['reduce'] = fuzz.trimf(oxygen_therapy.universe, [-1, -1, 0])
oxygen_therapy['maintain'] = fuzz.trimf(oxygen_therapy.universe, [-0.5, 0, 0.5])
oxygen_therapy['increase'] = fuzz.trimf(oxygen_therapy.universe, [0, 1, 1])

rules = [
    ctrl.Rule(heart_rate['low'] & blood_pressure['low'], [fluid_therapy['increase'], drug_dose['increase']]),
    ctrl.Rule(heart_rate['high'] & blood_pressure['high'], drug_dose['reduce']),
    ctrl.Rule(heart_rate['normal'] & blood_pressure['normal'], drug_dose['maintain']),
    ctrl.Rule(body_temp['high'] & oxygen_saturation['low'], [oxygen_therapy['increase'], drug_dose['increase']]),
    ctrl.Rule(body_temp['normal'] & oxygen_saturation['normal'], drug_dose['maintain']),
    ctrl.Rule(body_temp['low'] & oxygen_saturation['low'], [oxygen_therapy['increase'], fluid_therapy['increase']]),
    ctrl.Rule(blood_sugar['high'], drug_dose['increase']),
    ctrl.Rule(blood_sugar['low'], drug_dose['reduce']),
    ctrl.Rule(blood_sugar['normal'], drug_dose['maintain']),
]

system = ctrl.ControlSystem(rules)
simulation = ctrl.ControlSystemSimulation(system)
simulation.input['heart_rate'] = 50
simulation.input['blood_pressure'] = 85
simulation.input['body_temp'] = 38
simulation.input['oxygen_saturation'] = 88
simulation.input['blood_sugar'] = 150


print("Drug Dose Adjustment:", simulation.output['drug_dose'])
print("Fluid Therapy Adjustment:", simulation.output['fluid_therapy'])
print("Oxygen Therapy Adjustment:", simulation.output['oxygen_therapy'])

drug_dose.view(sim=simulation)
fluid_therapy.view(sim=simulation)
oxygen_therapy.view(sim=simulation)
plt.show()
```

```
import numpy as np
import skfuzzy as fuzz
import skfuzzy.control as ctrl
import matplotlib.pyplot as plt

food_type = ctrl.Antecedent(np.arange(0, 3, 1), 'food_type')
food_weight = ctrl.Antecedent(np.arange(0, 3, 1), 'food_weight')
initial_temp = ctrl.Antecedent(np.arange(0, 3, 1), 'initial_temp')
water_content = ctrl.Antecedent(np.arange(0, 3, 1), 'water_content')

cooking_time = ctrl.Consequent(np.arange(0, 3, 1), 'cooking_time')
heat_level = ctrl.Consequent(np.arange(0, 3, 1), 'heat_level')

food_type.automf(names=['soft', 'medium', 'hard'])
food_weight.automf(names=['light', 'medium', 'heavy'])
initial_temp.automf(names=['cold', 'warm', 'hot'])
water_content.automf(names=['low', 'medium', 'high'])

cooking_time.automf(names=['short', 'medium', 'long'])
heat_level.automf(names=['low', 'medium', 'high'])

rules = [
    ctrl.Rule(food_type['soft'] & food_weight['light'], cooking_time['short']),
    ctrl.Rule(food_type['hard'] | food_weight['heavy'], cooking_time['long']),
    ctrl.Rule(initial_temp['cold'], cooking_time['long']),
    ctrl.Rule(initial_temp['hot'], cooking_time['short']),
    ctrl.Rule(water_content['high'], heat_level['high']),
    ctrl.Rule(water_content['low'], heat_level['low']),
    ctrl.Rule(food_type['medium'], cooking_time['medium']),
    ctrl.Rule(food_weight['medium'], cooking_time['medium']),
    ctrl.Rule(initial_temp['warm'], cooking_time['medium'])
]


system = ctrl.ControlSystem(rules)
simulation = ctrl.ControlSystemSimulation(system)

simulation.input['food_type'] = 2
simulation.input['food_weight'] = 1
simulation.input['initial_temp'] = 0
simulation.input['water_content'] = 2

simulation.compute()

print("Cooking Time:", simulation.output['cooking_time'])
print("Heat Level:", simulation.output['heat_level'])

cooking_time.view(sim=simulation)
heat_level.view(sim=simulation)
plt.show()
```
